# here item_price value is total bill value so creating new column with same data for total bill and then we can find price of each item in new column

```python
df['total_price_order$'] = (df['item_price'].str.replace("$","")).astype(float)
```

# we can find single item price

```python
df['single_item_price$'] = df['total_price_order$'] / df['quantity']
df.head()
```
